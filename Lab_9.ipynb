```python
query = '''
select customer_id, count(rental_id) as may_count
from sakila.rentals_may
group by customer_id
order by customer_id'''
query_may =pd.read_sql_query(query, engine)
```

```python
query = '''
select customer_id, count(rental_id) as june_count
from sakila.rentals_june
group by customer_id
order by customer_id
'''
query_june = pd.read_sql_query(query,engine)
```

```python
data_mer = query_may.merge(query_june, how = "inner", left_on ='customer_id', right_on = 'customer_id')
```

```python
def summer_3(df):
    lst = []
    for i in range(len(df)):
        if df['may_count'][i] > df['june_count'][i]:
            lst.append("May Preference")
        elif df['may_count'][i] < df['june_count'][i]:
            lst.append("June Preference")
        else:
            lst.append("No Preference")
    df.insert(3,"Preference",lst)
    df.reset_index(inplace = True, drop=True)
    lst.clear()
    return df;
```

```python
summer_3(data_mer)
```
